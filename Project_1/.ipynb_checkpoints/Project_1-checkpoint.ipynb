```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification


df = pd.read_csv("data/titanic.csv", skiprows=1)
df.columns = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex',
              'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']

print(df.head())
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,2)
jitter = np.random.normal(0,0.05, len(df))
survival_jitter = df['Survived'] + jitter
plt.scatter(df['Age'], survival_jitter, alpha=0.6, s=30)
plt.xlabel('Age')
plt.ylabel('Survival')
plt.grid(True, alpha=0.3)

print(df.groupby('Survived')['Age'].describe())
```

```python
plt.subplot(1,2,2)
jitter = np.random.normal(0,0.05, len(df))
survival_jitter = df['Survived'] + jitter
plt.scatter(df['Sex'], survival_jitter, alpha=0.6, s=30)
plt.xlabel('Sex')
plt.ylabel('Survival')
plt.grid(True, alpha=0.3)

print(df.groupby('Survived')['Sex'].describe())
```

```python
plt.subplot(1,2,2)
jitter = np.random.normal(0,0.05, len(df))
survival_jitter = df['Survived'] + jitter
plt.scatter(df['Pclass'], survival_jitter, alpha=0.6, s=30)
plt.xlabel('Class')
plt.ylabel('Survival')
plt.grid(True, alpha=0.3)

print(df.groupby('Survived')['Pclass'].describe())
```

```python
dfclean = df.dropna(subset=['Embarked'])

plt.figure(figsize=(10,6))
plt.subplot(1,2,2)
jitter = np.random.normal(0,0.05, len(dfclean))
survival_jitter = dfclean['Survived'] + jitter
plt.scatter(dfclean['Embarked'], survival_jitter, alpha=0.6, s=30)
plt.xlabel('Port')
plt.ylabel('Survival')
plt.grid(True, alpha=0.3)
plt.show()

print(dfclean.groupby('Survived')['Embarked'].describe())
```

```python
plt.subplot(1,2,2)
jitter = np.random.normal(0,0.05, len(df))
survival_jitter = df['Survived'] + jitter
plt.scatter(df['Fare'], survival_jitter, alpha=0.6, s=30)
plt.xlabel('Fare')
plt.ylabel('Survival')
plt.grid(True, alpha=0.3)

print(df.groupby('Survived')['Fare'].describe())
```

After comparing the data using scatter plots to see the distribution of survivors over specific features, this model will
